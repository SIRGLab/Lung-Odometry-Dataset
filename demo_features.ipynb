# Demo

```python
import cv2
from matplotlib import pyplot as plt
from load_data import get_pair
from opencv_detectors import CVDetectors
from superpoint import SuperPoint
from loftr import LoFTR_Detector
```

```python
%matplotlib inline
```

## Load data example

```python
img1, img2, pose = get_pair('data/lung_example', 600, 615)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
print('Relative pose matrix:\n', pose)
```

## Initialize detectors

```python
sift = CVDetectors('sift') # Other arguments: 'akaze', 'brisk', 'brisk-freak', 'orb'
sp = SuperPoint()
loftr = LoFTR_Detector()
```

## Show matches

### SIFT

```python
sift.show_matches(img1, img2, undistort=True)
```

### SuperPoint

```python
sp.show_matches(img1, img2, undistort=True)
```

### LoFTR

```python
loftr.show_matches(img1, img2, undistort=True)
```
